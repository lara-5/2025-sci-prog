````python
import google.generativeai as genai
from google.colab import userdata

GOOGLE_API_KEY = userdata.get('GOOGLE_API_KEY')
genai.configure(api_key=GOOGLE_API_KEY)
````

````python
gemini_lite = genai.GenerativeModel('gemini-2.0-flash-lite')
gemini_think = genai.GenerativeModel('gemini-2.5-pro')
````

````python
import json, re

def optimize_prompt(prompt, output):
    instruction = f'''
    You are a prompt-optimizer. Improve the prompt given the model output.
    Return ONLY a JSON object with fields:
    - optimized_prompt
    - feedback
    - rationale

    Prompt: "{prompt}"
    Output: "{output}"
    '''

    response = gemini_think.generate_content(instruction)
    raw = response.text

    m = re.search(r"```json\n([\s\S]*?)\n```", raw)
    if m:
        data = json.loads(m.group(1))
    else:
        data = json.loads(raw)

    return data
````

````python
def run_optimization(prompt, iterations=5):
    current = prompt
    for i in range(1, iterations + 1):
        print(f"===== ITERATION {i} =====")
        print("Prompt:", current)

        lite_output = gemini_lite.generate_content(current).text
        print("Lite Output:", lite_output)

        result = optimize_prompt(current, lite_output)
        print("Feedback:", result['feedback'])
        print("Rationale:", result['rationale'])
        print("Optimized Prompt:", result['optimized_prompt'])
        print()

        current = result['optimized_prompt']

    return current
````

````python
initial_prompt = "Napiši esej o utjecaju glazbe na kreativnost."
run_optimization(initial_prompt, 5)
````

Ovaj proces olakšava stvaranje udaljenih asocijacija i neočekivanih kombinacija ideja, što je temelj kreativnosti.

    Dopamin, ključni neurotransmiter u sustavu nagrađivanja mozga, ima centralnu ulogu u ovom procesu. Istraživanja na životinjama pokazuju da dopamin ne samo da potiče motivaciju, već i smanjuje latentnu inhibiciju (tendenciju mozga da ignorira poznate podražaje), omogućavajući pojedincima da lakše prepoznaju nove veze između naizgled nepovezanih ideja. fMRI studije pokazuju povećanu aktivnost u područjima bogatim dopaminom (npr., striatum) tijekom slušanja glazbe koja pobuđuje emocije, što sugerira da glazba pojačava dopaminski signal, povećavajući kognitivnu fleksibilnost i potencijal za stvaranje inovativnih ideja.

2.  **Afektivni mehanizmi: Od emocije do uvida**

    Glazba generira širok raspon emocija, od sreće i uzbuđenja do tuge i melankolije. Ove emocije, ako su usmjerene i dozirane, mogu djelovati kao katalizatori kreativnosti. Predloženi neurokognitivni model objašnjava kako glazbom inducirane estetske emocije mogu potaknuti kreativni uvid.

    **Model:** Slušanje glazbe aktivira limbički sustav i sustav nagrađivanja mozga, što dovodi do oslobađanja neurotransmitera poput dopamina i endorfina. Estetski doživljaji, kao što je *frisson* (ježenje), posebno aktiviraju amigdalni kompleks i prefrontalni korteks. Emocionalni odgovor rezultira privremenim smanjenjem kognitivne kontrole prefrontalnog korteksa, otvarajući put nesvjesnim idejama da uđu u svijest. To stvara "aha" moment, nagli uvid u rješenje problema.

    Glazbom modulirano pozitivno raspoloženje, kao što je sugerirano u "broaden-and-build" teoriji, proširuje opseg pažnje, omogućujući pojedincima da primijete više detalja i odnose u okolini. Ovo povećava pristup širem repertoaru kognitivnih strategija za rješavanje problema. Pozitivne emocije potiču fleksibilnost razmišljanja i potiču istraživanje novih ideja, čime se stvara plodno tlo za kreativno stvaranje. Primjeri uključuju kompozitore koji koriste glazbu za generiranje inspiracije ili pisce koji slušaju glazbu da bi se osjećali više inspiriranim, što im omogućuje da razviju složene ideje.

3.  **Strukturna analogija i kognitivni transfer:**

    Glazba, kao apstraktna umjetnička forma, nudi izvanrednu strukturu koja se može transponirati u druge domene kreativnosti. Kognitivni mehanizmi koji omogućuju ovaj transfer uključuju mapiranje apstraktnih shema. Um mozga prepoznaje i obrađuje ritmičke obrasce, melodijske konture, harmonijske progresije i kompozicijske forme. Ti se principi mogu primijeniti u drugim područjima stvaranja.

    Primjeri uključuju:
    *   **Arhitektura:** Iannis Xenakis, kompozitor i arhitekt, koristio je matematičke i glazbene principe (npr. stohastičke procese i Fibonaccijev niz) u dizajnu zgrada, stvarajući inovativne i vizualno stimulirajuće strukture. Glazbene kompozicije su mu služile kao model za prostorne odnose i kompozicijske elemente.
    *   **Narativna struktura:** James Joyce u romanu *Uliks* koristio je glazbene tehnike poput leitmotiva, kontrapunkta i ritmičke varijacije kako bi strukturirao narativ, dajući mu glazbenu dimenziju i dubinu. Glazbeni koncepti transformirani su u strukture pripovijedanja, poboljšavajući inovativnost.
    *   **Algoritamska elegancija:** Korištenje glazbenih algoritama u razvoju softvera. Na primjer, matematičke strukture koje se koriste u skladanju (npr. zlatni rez) mogu se upotrijebiti za stvaranje učinkovitih algoritama u dizajnu softvera.

    Kognitivni transfer se temelji na sposobnosti mozga da prepozna i apstrahira strukture, bez obzira na medij.

4.  **Kontekstualni moderatori i granice utjecaja:**

    Utjecaj glazbe na kreativnost nije univerzalan; ovisi o nizu kontekstualnih čimbenika i individualnih razlika.
    *   **Vrsta glazbe:** Istraživanja sugeriraju da različite vrste glazbe imaju različite učinke. Instrumentalna glazba, osobito ona s niskim razinama složenosti i bez riječi, često je povezana s povećanom kreativnošću, jer ne zahtijeva kognitivni napor za obradu riječi. Žanrovi kao što su klasična glazba, ambijentalna glazba ili instrumentalni jazz često se koriste u kreativnim procesima. U isto vrijeme, kompleksna ili zbunjujuća glazba može imati suprotan učinak, ometajući kreativnost.
    *   **Složenost zadatka:** Utjecaj glazbe na kreativnost ovisi o vrsti zadatka. U jednostavnim zadacima, glazba može pružiti ugodnu pozadinu, dok u složenim zadacima može ometati pažnju. Primjerice, glazba s brzim tempom može pomoći u stimulaciji i koncentraciji za ponavljajuće zadatke, dok može biti ometajuća za zadatke koji zahtijevaju detaljno razmišljanje.
    *   **Individualne razlike:** Individualne karakteristike, kao što su glazbeno iskustvo, osobnost i afektivno stanje, također utječu na odgovor na glazbu. Ljudi s glazbenom naobrazbom mogu bolje koristiti glazbu kao inspiraciju, dok introverznije osobe mogu preferirati glazbu s nižim razinama stimulacije.
    *   **Obrnuta U-krivulja (Yerkes-Dodsonov zakon):** Ova hipoteza predviđa da postoji optimalna razina uzbuđenja (arousal) za postizanje najboljih rezultata. Prema tome, glazba može poboljšati kreativnost do određene razine uzbuđenja, ali prekomjerna stimulacija može dovesti do preopterećenja i smanjenja kreativnosti. Primjerice, u fazi generiranja ideja, glazba s umjerenim tempom i složenošću može biti korisna, dok bi u fazi evaluacije ideja, tiha, ambijentalna glazba ili tišina mogla biti učinkovitija.

    Proučavanje ovih kontekstualnih moderatora ključno je za primjenu glazbe kao alata za poticanje kreativnosti.

**Integracija primjera:**

Einsteinovo sviranje violine pruža zanimljivu studiju slučaja za razumijevanje odnosa između glazbe i kreativnosti. Iako se često spominje kao dokaz glazbene inspiracije, njegov primjer može poslužiti za dublju analizu. Sviranje violine, osim užitka, moglo je služiti kao metoda aktivacije DMN-a. Motorna aktivnost (sviranje instrumenta) i slušanje glazbe zajedno mogu dovesti do inkubacije, potičući ne samo razmišljanje, nego i generiranje novih ideja.

**Zaključak**

U ovom eseju pokazano je da glazba djeluje kao kompleksni modulator kreativnih procesa. Glazba utječe na kreativnost kroz neuronske mreže, afektivne mehanizme i kognitivni transfer. Optimalni utjecaj glazbe ovisi o kontekstualnim čimbenicima i individualnim karakteristikama.

Za optimizaciju kreativnih okruženja preporučuje se:
*   Uključivanje instrumentalne glazbe s umjerenom složenošću.
*   Doziranje glazbe i njeno prilagođavanje, ovisno o fazi kreativnog procesa (generiranje ideja vs. evaluacija).
*   Razmatranje individualnih sklonosti prema glazbi i glazbenoj naobrazbi.

Buduća istraživanja trebala bi se usredotočiti na personalizirane pristupe glazbi, gdje se glazba prilagođava individualnim potrebama. Hipoteza je da personalizirana glazba koja sinkronizira s alfa valovima mozga selektivno pojačava aktivnost DMN-a, što dovodi do mjerljivog porasta u divergentnom razmišljanju. Daljnja istraživanja trebala bi istražiti učinak različitih žanrova i glazbenih struktura na kreativne procese, uzimajući u obzir složenost zadatka i osobne karakteristike pojedinca. Razumijevanje složenog odnosa između glazbe i kreativnosti može dovesti do inovativnih alata i strategija za optimizaciju kreativnog potencijala u različitim domenama.
